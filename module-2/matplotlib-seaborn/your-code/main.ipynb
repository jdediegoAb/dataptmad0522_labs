```python
# Method 1
import matplotlib.pyplot as plt
plt.plot(x, y,label='x*2',color='green')
plt.plot(x, z,label='x**2',linestyle='dashed')
plt.legend()
plt.show()

#color='green', linestyle='dashed', linewidth = 3, marker='o', markerfacecolor='blue', markersize=12
```

#### Use your previous code but now, resize your plot.
**Hint**: Add the figsize argument in plt.subplots().

If you want, try to add a title to the plot or even axes labels. You can also play with the fontweight and fontsize of the titles and labels. 

```python
# your code here
fig,(x1, x2)=plt.subplots(nrows=1,ncols=2,figsize=(10,3))
fig.suptitle('Example of lines',fontsize=18,fontweight='bold')
x1.plot(x,y,label = '$y=x^2$')
x2.plot(x,z,label= '$y=e^x$',color='red')
fig.legend()
```

#### Now you will have to plot both $y=x^2$ and $y=e^x$ in the same plot. Create two subplots and use a normal scale in the first one and a logarithmic scale in the second one.
**Hint**: Use `set_xscale` and `set_yscale`.

```python
# your code here
fig,ax=plt.subplots(nrows=1,ncols=1,figsize=(10,3))
fig.suptitle('Example of lines',fontsize=18,fontweight='bold')
ax.plot(x,y,label = '$y=x^2$',)
ax2 = ax.twinx()
ax2.set_yscale('log')
ax2.plot(x,z,label= '$y=e^x$',color='red')
fig.legend()
```

#### Write a loop to plot 3 scatter plots of the following features:

* Minutes Lightly Active vs Steps    
* Minutes Very Active vs Steps    
* Minutes Sedentary vs Steps  

```python
# your code here
x_col = "Steps"
y_columns = ['Minutes Lightly Active','Minutes Fairly Active','Minutes Very Active']
for y_col in y_columns:

    figure = plt.figure
    ax = plt.gca()
    ax.scatter(fitbit[x_col], fitbit[y_col])
    ax.set_xlabel(x_col)
    ax.set_ylabel(y_col)
    ax.set_title("{} vs {}".format(x_col, y_col))
```
