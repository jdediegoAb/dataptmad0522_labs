```python
import sqlalchemy
from sqlalchemy import create_engine
from sqlalchemy import inspect
import pandas as pd


```

```python
conection_string = 'sqlite:///../publications.db'
conection_string
```

```python
engine=create_engine(conection_string)
```

```python
engine
```

```python
inspector=inspect(engine)
inspector.get_table_names()
```

```python

#AUTHORS:au_id con TITLEAUTHOR:au_id done
#TITLEAUTHOR:title_id con TITLE:title_id HACER ESTA
#TITLE:pub_id con PUBLISHER:pub_id done
var='''
SELECT authors.au_id AS 'AUTHOR ID',
authors.au_lname AS 'LAST NAME',
authors.au_fname AS 'FIRST NAME',
ti.title AS 'TITLE',
pu.pub_name as 'PUBLISHER'
FROM authors
INNER JOIN titleauthor t ON authors.au_id=t.au_id
INNER JOIN titles ti ON ti.title_id=t.title_id
INNER JOIN publishers pu ON pu.pub_id=ti.pub_id
'''
```

```python
df = pd.read_sql_query(var, engine)
```

```python
df
```

```python
var2='''
SELECT authors.au_id AS 'AUTHOR ID',
authors.au_lname AS 'LAST NAME',
authors.au_fname AS 'FIRST NAME',
pu.pub_name as 'PUBLISHER',
COUNT(authors.au_id) AS ' COUNT TITLES'
FROM authors
INNER JOIN titleauthor t ON authors.au_id=t.au_id
INNER JOIN titles ti ON ti.title_id=t.title_id
INNER JOIN publishers pu ON pu.pub_id=ti.pub_id
GROUP BY authors.au_id
'''
```

```python
df2 = pd.read_sql_query(var2, engine)
```

```python
var3='''
SELECT authors.au_id AS 'AUTHOR ID',
authors.au_lname AS 'LAST NAME',
authors.au_fname AS 'FIRST NAME',
SUM(sa.qty) AS 'TOTAL'
FROM authors
INNER JOIN titleauthor t ON authors.au_id=t.au_id
INNER JOIN sales sa ON sa.title_id=t.title_id
GROUP BY authors.au_id
ORDER BY 4 DESC
LIMIT 3
'''
```

```python
df3 = pd.read_sql_query(var3, engine)
```

```python
df3
```

```python
#Challenge 4 Best Selling Authors Ranking
var4='''
SELECT authors.au_id AS 'AUTHOR ID',
authors.au_lname AS 'LAST NAME',
authors.au_fname AS 'FIRST NAME',
SUM(nullif(sa.qty,'NaN')) AS 'TOTAL'
FROM authors
LEFT JOIN titleauthor t ON authors.au_id=t.au_id
LEFT JOIN sales sa ON sa.title_id=t.title_id
GROUP BY authors.au_id
'''
```

```python
df4 = pd.read_sql_query(var4, engine)
```

```python
df4['TOTAL'] = df4['TOTAL'].fillna(0)
```
